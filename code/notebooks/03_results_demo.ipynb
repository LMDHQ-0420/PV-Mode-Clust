```python
import sys, os, glob, json
sys.path.insert(0, os.path.abspath('..'))
os.chdir(os.path.abspath('..'))
import numpy as np, pandas as pd, matplotlib.pyplot as plt
```

```python
# 主表：各站 ACC/RMSE 汇总
rows=[json.load(open(p)) for p in glob.glob('results/eval_*.json')]
if rows:
    df=pd.DataFrame(rows); print(df.groupby('station')[['acc','rmse','r2']].mean())
```

```python
# 消融汇总条形图（含泄漏变体 E 的虚高精度）
p='results/ablation/summary.csv'
if os.path.exists(p):
    s=pd.read_csv(p)
    plt.figure(figsize=(8,4)); plt.bar(s['variant'], s['acc_mean'], yerr=s['acc_std'])
    plt.ylabel('ACC (%)'); plt.title('ablation A-G (E=leak shows inflated acc)'); plt.show()
    print(s[['variant','acc_mean','acc_std','delta_acc','notes']])
```

```python
# 图1 区制隶属 / 图2 专家权重 / 图3 订正对比
SID='station00'
for name,title in [('regime_membership','fig1 regime'),('expert_weight','fig2 expert'),('correction_compare','fig3 correction')]:
    fp=f'results/interpret/{name}_{SID}.csv'
    if os.path.exists(fp):
        d=pd.read_csv(fp); d.plot(figsize=(10,3), title=title); plt.show()
```
